7. A Write a program to perform element-wise multiplication of two NumPy
 arrays.

```python
import numpy as np

def multiply_arrays(array1, array2):
    """
    Performs element-wise multiplication of two NumPy arrays.

    Args:
        array1 (np.ndarray): The first NumPy array.
        array2 (np.ndarray): The second NumPy array.

    Returns:
        np.ndarray: A new NumPy array containing the element-wise product of array1 and array2.
                      Returns None if the arrays have incompatible shapes.
    """
    if array1.shape != array2.shape:
        print("Error: Arrays must have the same shape for element-wise multiplication.")
        return None
    return np.multiply(array1, array2)

if __name__ == '__main__':
    array1 = np.array([[1, 2, 3], [4, 5, 6]])
    array2 = np.array([[7, 8, 9], [10, 11, 12]])

    result_array = multiply_arrays(array1, array2)

    if result_array is not None:
        print("Array 1:")
        print(array1)
        print("\nArray 2:")
        print(array2)
        print("\nElement-wise product:")
        print(result_array)
```

8. A Create a line plot with multiple lines using Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for multiple lines
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)

# Create the plot
plt.figure(figsize=(10, 6)) # Adjust figure size as needed

# Plot multiple lines with labels and different styles
plt.plot(x, y1, label='sin(x)', linestyle='-', color='blue')
plt.plot(x, y2, label='cos(x)', linestyle='--', color='red')
plt.plot(x, y3, label='sin(x) * cos(x)', linestyle=':', color='green')

# Add title and labels
plt.title('Multiple Lines Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add legend
plt.legend()

# Add grid for better readability
plt.grid(True)

# Show the plot
plt.show()
```

9. A Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.

```python
import pandas as pd
import numpy as np

# Create a sample DataFrame
data = {'col1': np.random.randint(1, 20, 10),
        'col2': np.random.rand(10),
        'col3': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'C', 'A']}
df = pd.DataFrame(data)

# Define the threshold
threshold = 10

# Filter rows where 'col1' is greater than the threshold
filtered_df = df[df['col1'] > threshold]

# Print the original and filtered DataFrames
print("Original DataFrame:")
print(df)
print("\nFiltered DataFrame (col1 > {}):".format(threshold))
print(filtered_df)
```

10. Create a histogram using Seaborn to visualize a distribution

```python
    import seaborn as sns
    import matplotlib.pyplot as plt
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)

# Create a basic histogram
sns.histplot(data=data)
plt.title('Basic Histogram')
plt.show()

# Create a histogram with customized bins and color
sns.histplot(data=data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram with Custom Bins and Color')
plt.show()

# Create a histogram with KDE
sns.histplot(data=data, kde=True)
plt.title('Histogram with KDE')
plt.show()

# Load the iris dataset
iris = sns.load_dataset('iris')

# Create a histogram with hue for comparison
sns.histplot(data=iris, x='sepal_length', hue='species')
plt.title('Histogram with Hue')
plt.show()
```

13. Create a 3D scatter plot using Plotly.

```python
import plotly.graph_objects as go
import numpy as np

# Generate random data for the plot
np.random.seed(42)
x = np.random.randn(100)
y = np.random.randn(100)
z = np.random.randn(100)
colors = np.random.rand(100) # Assign random colors to points

# Create the scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=8,
        color=colors, # Set color of markers to the 'colors' array
        colorscale='Viridis', # Choose a colorscale
        opacity=0.8
    )
)])

# Customize the layout (optional)
fig.update_layout(
    title='3D Scatter Plot Example',
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    ),
    margin=dict(l=0, r=0, b=0, t=40)
)

# Display the plot
fig.show()
```
